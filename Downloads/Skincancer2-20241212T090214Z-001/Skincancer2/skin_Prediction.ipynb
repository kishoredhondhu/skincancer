```python
import tensorflow as tf
import numpy as np

# Load the saved ResNet model
model = tf.keras.models.load_model('/content/RestNet.h5')

class_names = ["benign", "malignant"]  # Update with your class names

def preprocess_image(image_path):
    """
    Preprocesses the image for ResNet50.
    - Resizes the image to 224x224.
    - Converts the image to a NumPy array.
    - Applies ResNet50 preprocessing.
    """
    # Load and resize the image
    image = tf.keras.preprocessing.image.load_img(
        image_path, target_size=(224, 224)
    )
    # Convert to NumPy array
    image_array = tf.keras.preprocessing.image.img_to_array(image)
    # Expand dimensions to make it batch-compatible
    image_array = tf.expand_dims(image_array, 0)
    # Preprocess for ResNet
    image_array = tf.keras.applications.resnet50.preprocess_input(image_array)
    return image_array

def predict_image(image_path):
    """
    Predicts the class of the given image using the ResNet model.
    - Preprocesses the image.
    - Predicts the probabilities.
    - Returns the predicted class and confidence score.
    """
    image_array = preprocess_image(image_path)
    predictions = model.predict(image_array)
    predicted_class_index = np.argmax(predictions[0])  # Get the class index
    predicted_class = class_names[predicted_class_index]  # Map to class name
    confidence = round(100 * np.max(predictions[0]), 2)  # Get confidence score
    return predicted_class, confidence

# Example usage
image_path = "/content/maligant.jpeg"
predicted_class, confidence = predict_image(image_path)
print("Predicted class:", predicted_class)
print("Confidence:", confidence)
```
